**Distribution plots**

```python
import seaborn as sns
% matplotlib inline
```

##Seaborn comes with built-in datasets

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
#distplot shows distribution values of a column 
sns.distplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'],kde=False)
```

```python
sns.distplot(tips['total_bill'],kde=False, bins=40)
```

```python
#jointplot combines 2 different distributions
sns.jointplot(x='total_bill',y='tip',data = tips)
```

```python
sns.jointplot(x='total_bill',y='tip',data = tips,kind ='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data = tips,kind ='reg')
```

```python
sns.jointplot(x='total_bill',y='tip',data = tips,kind ='kde')
```

```python
# pairplot() makes plots of all the columns in the dataframe
sns.pairplot(tips)
```

```python
sns.pairplot(tips,hue='sex')#hue is for categorical attribute
```

```python
sns.pairplot(tips,hue='sex',palette='coolwarm')
```

```python
#rugplot is similar to distplot
sns.rugplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'],kde=False)
```

```python
sns.kdeplot(tips['total_bill'])
```

##**Matrixplots**

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
tips.corr()
```

```python
tc = tips.corr()
```

```python
#heatmap shows magnitude of a phenomenon
sns.heatmap(tc)
```

```python
sns.heatmap(tc,annot=True,cmap ='coolwarm')
```

```python
flights.pivot_table(index ='month',columns='year',values='passengers')
```

```python
fp = flights.pivot_table(index ='month',columns='year',values='passengers')
sns.heatmap(fp,cmap='magma',linecolor='black',linewidths=3)
```

```python
#Cluster similar groups
sns.clustermap(fp)
```

```python
sns.clustermap(fp,cmap='coolwarm',standard_scale=1)
```

##**GRID**

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].unique()
```

```python
sns.pairplot(iris)
```

```python
sns.PairGrid(iris)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
g = sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
tips.head()
```

```python
g = sns.FacetGrid(data = tips,col = 'time', row ='smoker')
g.map(sns.distplot,'total_bill')
```

```python
g = sns.FacetGrid(data = tips,col = 'time', row ='smoker')
g.map(plt.scatter,'total_bill','tip')
```
